```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect('tips.db')
cursor = conn.cursor()
```

```python
#Just to look up on the table
pd.read_sql_query('''
SELECT * FROM Day
''', conn)
```

```python
#Just to look up on the table
pd.read_sql_query('''
SELECT * FROM Observation
''', conn)
```

```python
#1
pd.read_sql_query('''
SELECT * FROM Observation WHERE total_bill>50
''', conn)
```

```python
#2
pd.read_sql_query('''
SELECT AVG(tip) AS average_tip FROM Observation
''', conn)
```

```python
#3
pd.read_sql_query('''
SELECT MAX(total_bill) AS max_total_bill FROM Observation
''', conn)
```

```python
#4
pd.read_sql_query('''
SELECT day, SUM(size) AS total_people_served
FROM Observation
JOIN Day ON Observation.day_id = Day.day_id
GROUP BY Day.day
''', conn)
```

```python
#5
pd.read_sql_query('''
SELECT size, AVG(total_bill) AS average_total_bill
FROM Observation
GROUP BY size
''', conn)
```

```python
#6
pd.read_sql_query('''
SELECT Smoker.smoker, SUM(Observation.tip) AS total_tip_amount
FROM Observation
JOIN Smoker ON Observation.smoker_id = Smoker.smoker_id
GROUP BY Smoker.smoker
''', conn)
```

```python
#7
pd.read_sql_query('''
SELECT day, (tip / total_bill) * 100 AS tip_percentage
FROM Observation
JOIN Day ON Observation.day_id = Day.day_id
GROUP BY Day.day
''', conn)
```

# **<font color='red'> The task required to determine the average tip percentage (tip/total_bill) for each day. Minus 1 point </font>**

```python
#8
pd.read_sql_query('''
SELECT Observation.total_bill, Day.day, Time.time
FROM Observation
JOIN Day ON Observation.day_id = Day.day_id
JOIN Time ON Observation.time_id = Time.time_id
''', conn)
```

```python
#9
pd.read_sql_query('''
SELECT Sex.sex, COUNT(*) AS total_observations
FROM Observation
JOIN Sex ON Observation.sex_id = Sex.sex_id
GROUP BY Sex.sex
''', conn)
```

```python
#10
pd.read_sql_query('''
SELECT Observation.tip
FROM Observation
JOIN Day ON Observation.day_id = Day.day_id
JOIN Time ON Observation.time_id = Time.time_id
WHERE Day.day = 'Sat' AND Time.time = 'Dinner'
''', conn)
```

```python
#11
pd.read_sql_query('''
SELECT Sex.sex, MAX(Observation.total_bill) AS largest_total_bill
FROM Observation
JOIN Sex ON Observation.sex_id = Sex.sex_id
GROUP BY Sex.sex
''', conn)
```

```python
#12
pd.read_sql_query('''
SELECT AVG(tip) AS average_tip
FROM Observation
WHERE size >4
''', conn)
```

```python
#13
pd.read_sql_query('''
SELECT Observation.* , day, time
FROM Observation
JOIN Day ON Observation.day_id = Day.day_id
JOIN Time ON Observation.time_id = Time.time_id
WHERE Day.day IN ('Fri', 'Sat')
AND Time.time IN ('Lunch', 'Dinner')
''', conn)
```

```python
#14
pd.read_sql_query('''
SELECT *
FROM Observation
WHERE (tip / total_bill) * 100 > 20
''', conn)
```

15. Use a subquery (query inside query) to find the record with the second highest tip. (you need to use a single query)

```python
#Bonus
pd.read_sql_query('''
SELECT *
FROM Observation
WHERE tip = (
    SELECT MAX(tip)
    FROM Observation
    WHERE tip < (SELECT MAX(tip) FROM Observation)
)
''', conn)
```
