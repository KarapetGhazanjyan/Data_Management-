## Use the below code for requesting html info from the page.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd  # for working with data and creating a DataFrame

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36"
}

base_url = "https://www.imdb.com<placeholder>"
url = "https://www.imdb.com/chart/top/?ref_=nv_mv_250"
response = requests.get(url, headers=headers)
print(response.status_code)
```

```python
soup = BeautifulSoup(response.content, "html.parser")
all_list = []


li_tags = soup.find_all("li", class_="ipc-metadata-list-summary-item")
```

```python
for item in li_tags:

    title = item.find("h3", class_="ipc-title__text").get_text(strip=True)
    rank = title.split('.')[0]
    metadata = item.find_all("span", class_="cli-title-metadata-item")
    year = metadata[0].get_text(strip=True)
    duration = metadata[1].get_text(strip=True)
    age_rating = metadata[2].get_text(strip=True)
    star_rating_group = item.find("span", class_="ipc-rating-star--rating")
    star_rating = star_rating_group.get_text(strip=True)
    vote_number_group = item.find("span", class_="ipc-rating-star--voteCount")
    vote_number = vote_number_group.get_text(strip=True).strip("()")

    author_url = base_url.replace("<placeholder>", item.find("a")["href"])
    author_response = requests.get(author_url, headers=headers)
    author_soup = BeautifulSoup(author_response.content, 'html.parser')

    author_name = author_soup.find("a", class_="ipc-metadata-list-item__list-content-item ipc-metadata-list-item__list-content-item--link").get_text(strip=True)

    all_list.append((title, rank, year, duration, age_rating, star_rating, vote_number, author_name
    ))

df = pd.DataFrame(all_list, columns=["Title", "Rank Of Movies", "Year", "Duration", "Age Rating", "Star Rating", "Number of Voters", "Author Name"])
```
